```python
import requests
```

```python
def read_key(keyfile):
    with open(keyfile) as f:
        return f.readline().strip("\n")
```

```python
key = read_key("../key.txt")
type(key)
```

```python
school = requests.get("https://serpapi.com/search?engine=google_maps", params = {
    'engine': "google_maps",
    'location': 'United States',
    'q': "los angeles schools",
    'type': "search",
    'api_key': key,
    #'output': 'html'
}
)
```

```python
school = school.json()
```
